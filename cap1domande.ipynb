

1.   **Come defineresti "Machine Learning"?** Un algoritmo che, tramite un algoritmo, riesce a predirre un risultato dati dei parametri.
2.   **Puoi dirmi 4 posti dove brilla?** Spam detector, Immagini recognition, Predizione di dati, risolvere problemi dove non si riesce a trovare un algoritmo
3. **Cos'è un labeled training set?** Sono i risultati di determinati parametri
4. **QUali sono 2 comuni supervised tasks?** Regression e Classification
5. **Quali sono 2 comuni non unsupervised** Clustering e riduzione di dati
6. **Quale algoritmo utilizzeresti a un robot per imparare a camminare in un terreno sconosciuto?** UnSupervised learning e, in particolare, reinforcament learning 
7. **Quale tipo di algoritmo useresti per segmentare il tuo cliente in più gruppi?** Clustering
8. **Sceglieresti di fare lo spam detection come supervised o unsupervised?** Supervised
9. **Cos'è un online learning system?** Dividiamo il dataset in parti per far si che possa entrare, impara incrementalmente
10. **Cos'è out-of-core learning?** Divide il dataset in parti, che vengono poi iterate siccome il tutto non riesce ad entrare in memoria
11. **Quale algoritmo utilizza la similarità per le predizioni?** Instance based learning
12. **Qual'è la differenza fra modello parametro e hyperparametro?** Nel modello, si impara cambiando i pesi. L'iperparametro è uno dei parametro del modello
13. **Cosa cerca un model-based algoritmo? Qual'è la strategia più comune? Come fare delle predizioni?** Cerca di trovare l'algoritmo migliore per potere fare delle predizioni più accurate, e per trovarle lui cambia i parametri a seconda della cost function. Per fare le predizioni, dai gli input, lui li moltiplica per i parametri ed è fatto
14. **4 problemi del machine learning** I problemi principali sono riguardante i dati: pochi dati, dati non conformi. Questo causa problemi tipo underfitting e overfitting
15. **Se il tuo algoritmo va bene nel training ma non nel testing, cosa succede? Soluzioni?** Abbiamo fatto un overfitting. Per potere evitare, o semplifichiamo i nostri dati, oppure aumentiamo i nostri dati oppure diminuiamo i nostri disturbi
16. **Cos'è un test set? Come lo useresti?** Un test set è una parte del nostro dataset che, per vedere la vera accuratezza del nostro modello, noi poi facciamo  un training su di esso per vedere quanto bene sta andando
